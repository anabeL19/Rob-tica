```python
import math
import random
import numpy as np

from matplotlib import pyplot as plt
```

```python
class Modelo_pdf:
    def __init__(self):
        self.b = 1

    def distribucion_gaussiana(self, r):
        b = self.b
        suma = 0
        for i in range(r):
            suma += random.uniform(-b,b)
        return( 0.5*suma )   
        
    def distribucion_triangular(self):
        b = self.b
        return( math.sqrt(6)/2 * (random.uniform(-b,b) + random.uniform(-b,b)) )
    
    def pdf(self,muestras):
        dg = []
        dg20 = []
        dt = []
        for i in range (0, muestras):
            dg.append(self.distribucion_gaussiana(12))
            dg20.append(self.distribucion_gaussiana(20))
            dt.append(self.distribucion_triangular())
        return np.array(dg), np.array(dg20), np.array(dt)
```

## 10000 muestras 

```python
obj = Modelo_pdf()
muestras = 10000
dgaussiana, dgaussiana20, dtriangular = obj.pdf(muestras)
```

```python
plt.hist(dgaussiana, bins = 'auto')
plt.show()
```

```python
plt.hist(dgaussiana20, bins = 'auto')
plt.show()
```

```python
plt.hist(dtriangular, bins = 'auto')
plt.show()
```
